1. Является ли число 62,6738 рациональным или иррациональным? Почему?
ответ: рациональнoe

```python
from sympy import *
Rational(62.6738)

val = 62.6738
num = S(val)
print(f"{num} рациональное? {num.is_rational}")

simplified = nsimplify(val)
if simplified.is_irrational:
    print(f"Число {val} похоже на иррациональное: {simplified}")
else:
    print(f"Число {val} — рациональное: {simplified}")
```

```python
# 3. Вычислите 81**(1/2)
from sympy import *

x = symbols('x') 

f = 81**(x)
print(f"Функция: {sympify("81**(1/2)",evaluate=False)}")
print(f"Ответ: {f.evalf(1, subs={x: 1/2})}")
```

```python
'''
5. Каков будет остаток по кредиту в 1 000 долларов сроком на 3 года под 5 %
годовых, если проценты начисляются ежемесячно, а платежей в течение
этого периода нет?
'''

from sympy import *

amount = 1000
percent = 0.05
years = 3

rest = amount*(1 + percent/12)**(years * 12)

print(f"Ответ: {rest}")
```

```python
'''
6. Выполните предыдущее упражнение при условии, что проценты начисля-
ются непрерывно.
'''
from sympy import *
from math import exp

amount = 1000
percent = 0.05
years = 3
partitional = 365

rest = amount*(1 + percent/partitional)**(partitional * years)

print(f"Ответ: {rest}")

a = amount * exp(percent*years)

print(f"Ответ (exp): {a}")
```

```python
'''
7. Чему равен уклон графика функции при 3*x**2 + 1 при x = 3?
'''
from sympy import *
# Объявляем символ 'x' для SymPy
x = symbols('x')
# Теперь объявляем функцию через обычный синтаксис Python
f = 3*x**2
# Вычисляем производную функции
dx_f = diff(f)
print(f"Производная {dx_f}") # выводит 6*x
print(f"Ответ: {dx_f.subs(x,3)}") 
```

```python
'''
8. Чему равна площадь под графиком функции 3*x**2 + 1 в интервале от
 0 до 2 ?
'''
from sympy import *
# Объявляем символ 'x' для SymPy
x = symbols('x')
# Теперь объявляем функцию через обычный синтаксис Python
f = 3*x**2 + 1

# Вычисляем интеграл от функции по x в интервале от x = 0 до x = 1
area = integrate(f, (x, 0, 2))
print(f"Ответ: {area}") 
```
