1. Является ли число 62,6738 рациональным или иррациональным? Почему?
ответ: рациональнoe

```python
from sympy import *
Rational(62.6738)

val = 62.6738
num = S(val)
print(f"{num} рациональное? {num.is_rational}")

simplified = nsimplify(val)
if simplified.is_irrational:
    print(f"Число {val} похоже на иррациональное: {simplified}")
else:
    print(f"Число {val} — рациональное: {simplified}")
```

```python
# 2. Вычислите значение выражения 10**7 * 10**(-5)

from sympy import *
from sympy.abc import a, b

print((10**7  * 10**(-5)))

x, y = symbols('x y') 
f = 10**x * 10**y

print(f)
print(f.evalf(subs={x: 7,y:-5}))

sympify('10**a * 10**b',evaluate = False)
```

```python
# 3. Вычислите 81**(1/2)
from sympy import *

x = symbols('x') 

f = 81**(x)
print(f"Функция: {sympify("81**(1/2)",evaluate=False)}")
print(f"Ответ: {f.evalf(1, subs={x: 1/2})}")
```
